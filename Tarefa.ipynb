```python
import pandas as pd
file_data = pd.read_csv('arquivoTarefa.csv', header=None) 
print(file_data)
num_of_records = len(file_data)
print(num_of_records)
```

```python
records = []
for i in range(0,num_of_records):
    records.append([str(file_data.values[i,j]) for j in range(0,4)])
```

```python
import numpy as np
from numpy import nan

transactions=[]
for record in records:
    new_record = []
    for item in record:
        if str(item) != 'nan':
            new_record.append(item)
    print(new_record)
    transactions.append(new_record)
print(transactions)

from efficient_apriori import apriori
itemsets, rules = apriori(transactions, min_support=0.2, min_confidence=0.6)
```
